CREACIÓN DE TENSORES 

```python
import torch

# 1. Desde una lista de Python
tensor_lista = torch.tensor([[1, 2], [3, 4]])

# 2. Tensores de ceros, unos o aleatorios (muy útiles para inicializar pesos)
ceros = torch.zeros((2, 3))       # Matriz de 2 filas y 3 columnas llena de ceros
aleatorio = torch.randn((3, 3))   # Matriz 3x3 con distribución normal

print(f"Forma (shape) del tensor aleatorio: {aleatorio.shape}")
print(f"Tipo de dato: {aleatorio.dtype}")
```

OPERACIONES MATEMÁTICAS

```python
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[1, 1], [1, 1]])

# Suma elemento a elemento
suma = A + B 

# Multiplicación matricial (¡CRÍTICO EN REDES NEURONALES!)
# Se puede usar el operador @ o torch.matmul
mult_mat = A @ B
```

REDIMENSIONAMIENTO : RESHAPE/VIEW

```python
# Tensor de 1 dimensión con 6 elementos
x = torch.tensor([1, 2, 3, 4, 5, 6])

# Cambiar la forma a una matriz de 2x3
x_matriz = x.view(2, 3) 
# También puedes usar x.reshape(2, 3)

# El -1 es un comodín muy usado en exámenes. Le dice a PyTorch: 
# "Calcula esta dimensión automáticamente en base a las demás".
x_auto = x.view(3, -1) # PyTorch sabe que el -1 debe ser un 2.
```
