FUNCIONES DE PERDIDA Y OPTIMIZADORES 

Funcion de perdida -> Es lo que mide qué tan mal lo está haciendo el modelo.
El objetivo del entrenamiento es minimizar esa pérdida.



| Problema                 | Función de pérdida        |
| ------------------------ | ------------------------- |
| Regresión                | MSE (`MSELoss`)           |
| Regresión con outliers   | MAE (`L1Loss`)            |
| Clasificación binaria    | BCE (`BCEWithLogitsLoss`) |
| Clasificación multiclase | CrossEntropy              |


La función de pérdida se selecciona en función de la naturaleza del problema. En regresión se utilizan métricas como MSE para medir error continuo, mientras que en clasificación se emplea entropía cruzada para modelar probabilidades y penalizar predicciones incorrectas con alta confianza

-------------


Optimizador , es el algoritmo que toma los gradientes y actualiza los pesos de la red -> torch.optim

SGD (Stochastic Gradient Descent): El clásico. Lento pero seguro.

Adam: El más usado hoy en día porque adapta la velocidad de aprendizaje (learning rate) automáticamente. Es como un SGD inteligente.

Para crear un optimizador, hay que pasarle dos cosas: qué pesos tiene que actualizar (los de tu modelo) y el Learning Rate (la tasa de aprendizaje: qué tan grandes son los pasos que da para corregir el error).



```python
import torch.optim as optim
import torch.nn as nn

class IrisNet(nn.Module):
    def __init__(self): 
        super().__init__() 
        
        # Entran 4 características, salen 8 neuronas a la capa oculta
        self.fc1 = nn.Linear(in_features=4, out_features=8) 
        self.activacion = nn.ReLU()
        # Entran las 8 neuronas de la capa oculta, salen las 3 clases finales
        self.fc2 = nn.Linear(in_features=8, out_features=3) 

    def forward(self, x):
        x = self.fc1(x)
        x = self.activacion(x)
        x = self.fc2(x)
        return x
```

```python

modelo = IrisNet()

# Definimos Función de pérdida para clasificación
criterion = nn.CrossEntropyLoss()

# Definimos el optimizador (Le pasamos los parámetros del modelo y un learning rate)
optimizer = optim.Adam(modelo.parameters(), lr=0.01)
```
