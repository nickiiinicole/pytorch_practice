Construcción de modelos torch.nn:

1. La anatomía de un modelo en PyTorch
Todo modelo en PyTorch debe heredar de la clase maestra nn.Module. Una red siempre tiene dos partes obligatorias:

__init__(self): Defines qué capas vas a usar, pero no cómo se conectan.

**forward**(self, x):como viaja el dato x a través de las piezas que guardaste en el __init__.





Las Capas Lineales y las Funciones de Activación

nn.Linear(in_features, out_features): Es la capa clásica "Totalmente Conectada" (Dense en Keras/TensorFlow). Adentro, simplemente hace $y = xW^T + b$. Necesitas decirle cuántos datos le entran y cuántos salen.Funciones de Activación (nn.ReLU(), nn.Sigmoid()...): Son las que rompen la linealidad. Sin ellas, una red neuronal gigante de 100 capas equivaldría matemáticamente a una sola capa lineal, y no podría aprender cosas complejas.

La función de activación se selecciona en función del rango y la interpretación de la variable objetivo. En regresión se requiere salida continua sin restricción, mientras que en clasificación se necesitan probabilidades, lo que justifica el uso de Sigmoid o Softmax

| Tipo de problema         | Activación final                       |
| ------------------------ | -------------------------------------- |
| Regresión                | Ninguna (lineal)                       |
| Clasificación binaria    | Sigmoid                                |
| Clasificación multiclase | Softmax (o ninguna + CrossEntropyLoss) |


```python
import torch.nn as nn

class FirstNet(nn.Module):
    def __init__(self):
        super().__init__() 
        
        self.capa_oculta = nn.Linear(in_features=10, out_features=5)
        self.activacion = nn.ReLU()
        self.capa_salida = nn.Linear(in_features=5, out_features=1)
        
    def forward(self, x):
        x = self.capa_oculta(x)
        x = self.activacion(x)
        x = self.capa_salida(x)
        return x # Devolvemos la predicción
```

EJERCICIO BASICO PRACTICO

El dataset Iris tiene 4 características de entrada (longitud y ancho de pétalos y sépalos) y queremos clasificar entre 3 tipos de flores (clases de salida).

Sabiendo esto, ¿te animas a escribir el código para crear una clase llamada RedIris que herede de nn.Module, contenga una capa oculta de 8 neuronas con función de activación ReLU, y finalice en una capa de salida adecuada para este problema?

```python

class IrisNet(nn.Module):
    def __init__(self): 
        super().__init__() 
        
        # Entran 4 características, salen 8 neuronas a la capa oculta
        self.fc1 = nn.Linear(in_features=4, out_features=8) 
        self.activacion = nn.ReLU()
        # Entran las 8 neuronas de la capa oculta, salen las 3 clases finales
        self.fc2 = nn.Linear(in_features=8, out_features=3) 

    def forward(self, x):
        x = self.fc1(x)
        x = self.activacion(x)
        x = self.fc2(x)
        return x
```
